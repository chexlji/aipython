```python
#붓꽃 데이터 읽어와서 확인하기
#1.데이터파일 읽어오기
import pandas as pd
df=pd.read_csv('iris.csv')
#2. 데이터의 기본정보 출력
df.info()
#3. 데이터의 기초 통계량 출력
print(df.describe())
#4. 품종별 개수 구하기
species_counts = df['species'].value_counts()
print(species_counts)
```

```python
#붓꽃 데이터 전처리
#1. 결측치 학인하기
df.isnull().sum()
#2. 중복 데이터 확인하기
df[df.duplicated]
#3. 중복 데이터 모두 확인하기
idx=(df['sepal_length']==5.8) & (df['petal_width']==1.9)
df.loc[idx,:]
```

```python
#4. 중복 데이터 삭제하기
df2=df.drop_duplicates()
```

```python
#붓꽃 데이터 그룹핑
#1. 품종(species) 열을 기준으로 합계 구하기
species_sum = df.groupby('species').sum()
print(species_sum)
#2. 품종(species) 열을 기준으로 평균 구하기
species_mean = df.groupby('species').mean()
print(species_mean)
```

```python
#붓꽃 데이터 시각화
#1. 막대 그래프 그리기
  #1. 꽃받침 길이의 시각화
import matplotlib.pyplot as plt
plt.bar(df.index[:10], df.sepal_length[:10])
```

```python
  #2. 붓꽃 종류별 꽃받침 길이의 평균에 대한 시각화
plt.bar(species_mean.index, species_mean['sepal_length'])
#2. 히스토그램 그래프 그리기
  #• 각 특징(4가지) 값들의 빈도수에 대한 시각화
df.hist()
#3. 산점도 그래프 그리기
  #• 꽃받침의 길이와 꽃잎의 길이의 산점도를 시각화
plt.scatter(df['sepal_length'], df['petal_length'])
```
