#0~100 점수를 입력받아서 90점 이상 A 80점 이상 B 70점이상 C 60점이상 D 그외 F

```python
score=int(input("0~100 점수를 입력하시오."))
if score>=90:
  grade= 'A'
elif score>=80:
  grade= 'B'
elif score>=70:
  grade= 'C'
elif score>=60:
  grade= 'D'
else:
  grade= 'F'
print(grade)
```
